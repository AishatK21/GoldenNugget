```python
import pandas as pd
import numpy as np
from pandas import DataFrame
```

```python
df = DataFrame()
```

```python
df = pd.read_csv('Datathon/USDA_PDP_AnalyticalResults.csv', low_memory=False)
```

```python
df
```

```python
# Create DataFrames for different states based on the first two letters of 'Sample ID'

# Filter rows where 'Sample ID' starts with 'NY' (New York)
ny_data = df[df['Sample ID'].str.startswith('NY')]

# Filter rows where 'Sample ID' starts with 'CA' (California)
ca_data = df[df['Sample ID'].str.startswith('CA')]

# Filter rows where 'Sample ID' starts with 'TX' (Texas)
tx_data = df[df['Sample ID'].str.startswith('TX')]

# Display the first few rows of each state's dataset to verify
print("New York Data:")
display(ny_data.head())

print("\nCalifornia Data:")
display(ca_data.head())

print("\nTexas Data:")
display(tx_data.head())
```

```python
# Get the unique states from the 'Sample ID' column by extracting the first two characters
unique_states = df['Sample ID'].str[:2].unique()

# Create a dictionary to store DataFrames for each state
state_data = {}

# Loop through each state code and create a separate DataFrame
for state in unique_states:
    # Filter the rows where the 'Sample ID' starts with the state code
    state_data[state] = df[df['Sample ID'].str.startswith(state)]
    
    # Display the first few rows of each state's dataset (optional)
    print(f"\nData for state: {state}")
    display(state_data[state].head())
```

```python
# Create a new column for the state based on the first two characters of 'Sample ID'
df['State'] = df['Sample ID'].str[:2]

# Group the data by 'State' and 'Pesticide Name' and calculate the mean concentration
average_concentration = df.groupby(['State', 'Pesticide Name'])['Concentration'].mean().reset_index()

# Display the result
print(average_concentration)
```

```python
# Assuming you have already created the 'average_concentration' DataFrame
# You can filter the result by state, for example, 'CA' for California

state = 'AL'  # Change 'CA' to the desired state code (e.g., 'NY', 'TX', etc.)

# Filter the DataFrame to show only the rows for the selected state
state_avg_concentration = average_concentration[average_concentration['State'] == state]

# Sort by 'Concentration' in descending order
state_avg_concentration_sorted = state_avg_concentration.sort_values(by='Concentration', ascending=False)

# Display the result for the selected state, sorted by highest to lowest average concentration
print(f"Average concentration of pesticides in {state} (sorted by highest concentration):")
display(state_avg_concentration_sorted)
```

```python
# Assuming you have already created the 'average_concentration' DataFrame

# Find the row with the highest concentration for each state
highest_concentration_per_state = average_concentration.loc[average_concentration.groupby('State')['Concentration'].idxmax()]

# Sort the result by state for clarity (optional)
highest_concentration_per_state = highest_concentration_per_state.sort_values(by='State')

# Display the result
print("Pesticide with the highest concentration for each state:")
display(highest_concentration_per_state)
```

```python
# Assuming 'highest_concentration_per_state' DataFrame has been created

# Save the DataFrame to a CSV file
highest_concentration_per_state.to_csv('highest_concentration_per_state.csv', index=False)

# Message to confirm the file has been saved
print("The data has been saved to 'highest_concentration_per_state.csv'.")
```
