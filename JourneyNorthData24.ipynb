```python
# Step 1: Import pandas and numpy
import numpy as np
import pandas as pd

# Step 2: Load the CSV into a DataFrame
df = pd.read_csv('Journey North Data.csv')

# Step 3: View the first few rows of the dataframe to ensure it's loaded correctly
df.head()
```

```python
df
```

```python
print(df.columns)
```

```python
# Step 1: Group the data by 'State/Province' and calculate the mean of 'Number' (sightings)
df_grouped = df.groupby('State/Province', as_index=False)['Number'].mean()

# Step 2: Rename the column to indicate it's an average
df_grouped.rename(columns={'Number': 'Average_Sightings'}, inplace=True)

# Step 3: View the updated dataframe
df_grouped
```

```python
# Save the dataframe to a CSV file
df_grouped.to_csv('average_butterfly_sightings.csv', index=False)
```
